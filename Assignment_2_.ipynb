Q3.
1. Sample Space ($\Omega$) : Let $\Omega = \{\omega_0, \omega_1, \omega_2, \omega_3\}$, where $\omega_k$ represents the event of $k$ transformer failures in a day.

$\sigma$-algebra ($\mathfrak{F}$): The power set $\mathfrak{P}(\Omega)$, containing all possible subsets of $\Omega$

Probability Measure ($P$): Defined by the given probabilities: $P(\{\omega_0\}) = 0.50$, $P(\{\omega_1\}) = 0.30$, $P(\{\omega_2\}) = 0.15$, $P(\{\omega_3\}) = 0.05$.


2. Random Variable X
$X$ is a function mapping from the sample space to the real numbers, $X: \Omega \to \mathbb{R}$, defined as:$$X(\omega_k) = k \quad \text{for } k \in \{0, 1, 2, 3\}$$

3. PMF
$$p_X(x) = \begin{cases} 0.50 & \text{if } x=0 \\ 0.30 & \text{if } x=1 \\ 0.15 & \text{if } x=2 \\ 0.05 & \text{if } x=3 \\ 0 & \text{otherwise} \end{cases}$$

4. $$E[X] = \sum x \cdot P(X=x)$$

5. $$E[X^2] = \sum x^2 \cdot P(X=x) $$
$\text{Var}(X) = E[X^2] - (E[X])^2$

6. Probabilities $P(X \ge 1)$ and $P(X \ge 2)$$P(X \ge 1) = 1 - P(X = 0) = 1 - 0.50 = 0.50$$P(X \ge 2) = P(X = 2) + P(X = 3) = 0.15 + 0.05 = 0.20$


7. Practical Meaning

Expectation (0.75): On average, the district can expect 0.75 transformer failures per day over a long period.

Variance (0.7875): This represents the day-to-day volatility or spread of the failures around the average.

```python
#Q3
import numpy as np
import matplotlib.pyplot as plt

print(" Q3: Transformer Failures \n")

# Given data
X_vals = np.array([0, 1, 2, 3])
P_X = np.array([0.50, 0.30, 0.15, 0.05])

E_X = np.sum(X_vals * P_X)
print(f"\n4. Expectation E[X] = {E_X:.2f} failures/day")


E_X_sq = np.sum((X_vals**2) * P_X)
Var_X = E_X_sq - (E_X**2)
print(f"5. Variance Var(X)  = E[X^2] - (E[X])^2 = {E_X_sq:.2f} - {E_X**2:.4f} = {Var_X:.4f}")
```

Determine the value of the maximum time $b$From the standard PERT model, the expectation is calculated as:$$E[X] = \frac{a + 4c + b}{6}$$We are given that the expected time must perfectly match the most likely time (the mode). Therefore, we set $E[X] = c$:$$c = \frac{a + 4c + b}{6}$$Now, let's solve for $b$ algebraically:$$6c = a + 4c + b$$$$2c = a + b$$$$b = 2c - a$$This derived equation ($2c = a + b$) is quite elegant. It tells us that for the mean to equal the mode in a PERT distribution, the mode $c$ must sit exactly halfway between $a$ and $b$.Given $a = 5$ and $c = 11$, plug them in to find $b$:$$b = 2(11) - 5 = 22 - 5 = 17$$The maximum time $b = 17$ days.

2. What happens to the variance?Let's look at the shape parameters first. Because $c$ is exactly in the middle of $a$ and $b$, the distances are equal ($c-a = 6$ and $b-c = 6$). If we plug these into the PERT shape parameter formulas from Q6:$\alpha = 1 + 4(\frac{6}{12}) = 1 + 2 = 3$$\beta = 1 + 4(\frac{6}{12}) = 1 + 2 = 3$Because $\alpha = \beta = 3$, the distribution becomes perfectly symmetric. It looks like a classic bell curve bound between 5 and 17.To find the actual numerical variance, we can plug our numbers into the approximation formula:$$\text{Var}(X) \approx \frac{(E[X] - a)(b - E[X])}{7}$$$$\text{Var}(X) = \frac{(11 - 5)(17 - 11)}{7} = \frac{(6)(6)}{7} = \frac{36}{7} \approx 5.14$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

print("=== Q7: PERT Expectation Matching Intuition ===\n")

# Given parameters
a = 5
c = 11

# 1. Calculate maximum time 'b'
b = 2 * c - a
print(f"1. Maximum time b:")
print(f"   Since E[X] = c, we get c = (a + 4c + b) / 6")
print(f"   Solving for b gives b = 2c - a")
print(f"   b = 2({c}) - {a} = {b} days\n")

# 2. Calculate Variance and show shape parameters
alpha = 1 + 4 * ((c - a) / (b - a))
beta_param = 1 + 4 * ((b - c) / (b - a))

E_X = c # Given by the problem condition
Var_X = ((E_X - a) * (b - E_X)) / 7

print(f"2. What happens to the variance?")
print(f"   Shape parameters become equal: α = {alpha:.0f}, β = {beta_param:.0f}")
print(f"   The distribution becomes perfectly symmetric.")
print(f"   Var(X) = ({E_X} - {a}) * ({b} - {E_X}) / 7 = 36/7 ≈ {Var_X:.4f}")

# Visualization: Symmetric PERT

x = np.linspace(a, b, 500)
x_scaled = (x - a) / (b - a)
pert_pdf = beta.pdf(x_scaled, alpha, beta_param) / (b - a)

plt.figure(figsize=(10, 5))
plt.plot(x, pert_pdf, 'purple', lw=2, label=f'PERT(a={a}, b={b}, c={c})')
plt.fill_between(x, pert_pdf, alpha=0.3, color='purple')

# Mark the center where Mean and Mode overlap
plt.axvline(E_X, color='black', linestyle='--', lw=2, label=f'Mean & Mode = {E_X}')

plt.title('Perfectly Symmetric PERT Distribution')
plt.xlabel('Completion Time (Days)')
plt.ylabel('Probability Density')
plt.xlim(3, 19)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
