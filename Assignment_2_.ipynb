Q3.
1. Sample Space ($\Omega$) : Let $\Omega = \{\omega_0, \omega_1, \omega_2, \omega_3\}$, where $\omega_k$ represents the event of $k$ transformer failures in a day.

$\sigma$-algebra ($\mathfrak{F}$): The power set $\mathfrak{P}(\Omega)$, containing all possible subsets of $\Omega$

Probability Measure ($P$): Defined by the given probabilities: $P(\{\omega_0\}) = 0.50$, $P(\{\omega_1\}) = 0.30$, $P(\{\omega_2\}) = 0.15$, $P(\{\omega_3\}) = 0.05$.


2. Random Variable X
$X$ is a function mapping from the sample space to the real numbers, $X: \Omega \to \mathbb{R}$, defined as:$$X(\omega_k) = k \quad \text{for } k \in \{0, 1, 2, 3\}$$

3. PMF
$$p_X(x) = \begin{cases} 0.50 & \text{if } x=0 \\ 0.30 & \text{if } x=1 \\ 0.15 & \text{if } x=2 \\ 0.05 & \text{if } x=3 \\ 0 & \text{otherwise} \end{cases}$$

4. $$E[X] = \sum x \cdot P(X=x)$$

5. $$E[X^2] = \sum x^2 \cdot P(X=x) $$
$\text{Var}(X) = E[X^2] - (E[X])^2$

6. Probabilities $P(X \ge 1)$ and $P(X \ge 2)$$P(X \ge 1) = 1 - P(X = 0) = 1 - 0.50 = 0.50$$P(X \ge 2) = P(X = 2) + P(X = 3) = 0.15 + 0.05 = 0.20$


7. Practical Meaning

Expectation (0.75): On average, the district can expect 0.75 transformer failures per day over a long period.

Variance (0.7875): This represents the day-to-day volatility or spread of the failures around the average.

```python
#Q3
import numpy as np
import matplotlib.pyplot as plt

print(" Q3: Transformer Failures \n")

# Given data
X_vals = np.array([0, 1, 2, 3])
P_X = np.array([0.50, 0.30, 0.15, 0.05])

E_X = np.sum(X_vals * P_X)
print(f"\n4. Expectation E[X] = {E_X:.2f} failures/day")


E_X_sq = np.sum((X_vals**2) * P_X)
Var_X = E_X_sq - (E_X**2)
print(f"5. Variance Var(X)  = E[X^2] - (E[X])^2 = {E_X_sq:.2f} - {E_X**2:.4f} = {Var_X:.4f}")
```

1. Sample Space ($\Omega$): The interval $[0, 1]$.$\sigma$-algebra ($\mathfrak{F}$): The Borel $\sigma$-algebra on $[0, 1]$, denoted as $\mathcal{B}([0, 1])$.Probability Measure ($P$): Defined by the integral of the probability density function over any Borel set $A \in \mathfrak{F}$:$$P(A) = \int_A f(x) \,dx = \int_A \frac{1}{B(2,18)} x(1-x)^{17} \,dx$$

2. $X$ is the identity function mapping the sample space to the real numbers:$$X: \Omega \to \mathbb{R} \quad \text{where} \quad X(\omega) = \omega$$

3. Given $\alpha = 2$ and $\beta = 18$, the pdf is:$$f_X(x) = \begin{cases} \frac{1}{B(2,18)} x^1 (1-x)^{17}, & 0 < x < 1 \\ 0, & \text{otherwise} \end{cases}$$

4. To be a valid pdf,
1) Non-negativity: For $0 < x < 1$, both $x > 0$ and $(1-x) > 0$. Since the Beta function $B(2,18)$ is a positive constant, $f(x) \ge 0$ everywhere.
2) Integrates to 1: By the very definition of the Beta function, $B(\alpha, \beta) = \int_0^1 x^{\alpha-1}(1-x)^{\beta-1} \,dx$. Therefore:$$\int_0^1 f(x) \,dx = \frac{1}{B(2,18)} \int_0^1 x^{2-1}(1-x)^{18-1} \,dx = \frac{B(2,18)}{B(2,18)} = 1$$

5. Expectation $E[X]$The expected value of a Beta distribution is given by the formula $\frac{\alpha}{\alpha + \beta}$.$$E[X] = \frac{2}{2 + 18} = \frac{2}{20} = 0.10$$

6. Variance $\text{Var}(X)$The variance of a Beta distribution is given by $\frac{\alpha\beta}{(\alpha+\beta)^2(\alpha+\beta+1)}$.$$\text{Var}(X) = \frac{2 \times 18}{(2 + 18)^2 (2 + 18 + 1)} = \frac{36}{(20)^2(21)} = \frac{36}{400 \times 21} = \frac{36}{8400} = \frac{3}{700} \approx 0.004286$$

7. Leaving this in Beta integral form (also known as the Incomplete Beta Function):$$P(X \le 0.10) = \int_0^{0.10} \frac{1}{B(2,18)} x(1-x)^{17} \,dx$$

8. Probability that the defective proportion lies between 0.05 and 0.15Leaving this in integral form:$$P(0.05 \le X \le 0.15) = \int_{0.05}^{0.15} \frac{1}{B(2,18)} x(1-x)^{17} \,dx$$

9. The Beta distribution is specifically defined on the interval $[0, 1]$, making it the mathematically natural choice for modeling proportions, percentages, or probabilities. Furthermore, its shape parameters ($\alpha$ and $\beta$) allow it to flexibly model skewed real-world data, such as a factory where the defect rate is typically low and heavily skewed toward $0$.

10. Expectation (0.10): On an average day, exactly 10% of the production batch will be defective.Variance ($\approx 0.0043$): This represents the day-to-day volatility of the manufacturing process.
