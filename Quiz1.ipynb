```python
#Q1
import math

print("=== Q1: Probability of Shooting at a Target ===\n")

# Given parameters
R_outer = 10
R_inner = 9  # Within 1 unit of the boundary means between radius 9 and 10

# ---------------------------------------------------------
# Model 1: Uniformly distributed with respect to AREA
# Probability = (Area of outer ring) / (Total Area)
# ---------------------------------------------------------
area_total = math.pi * (R_outer ** 2)
area_inner = math.pi * (R_inner ** 2)
area_ring = area_total - area_inner

prob_area = area_ring / area_total

print("Model 1 (Uniform by Area):")
print(f"- Total Area: {area_total:.2f}")
print(f"- Target Ring Area: {area_ring:.2f}")
print(f"- Exact Probability = {prob_area:.2f} (or {prob_area * 100:.0f}%)\n")

# ---------------------------------------------------------
# Model 2: Uniformly distributed with respect to RADIUS
# Probability = (Length of target radius interval) / (Total radius length)
# ---------------------------------------------------------
radius_total_length = R_outer - 0
target_radius_length = R_outer - R_inner

prob_radius = target_radius_length / radius_total_length

print("Model 2 (Uniform by Radius):")
print(f"- Total Radius Range: [0, {R_outer}] (Length: {radius_total_length})")
print(f"- Target Radius Range: [{R_inner}, {R_outer}] (Length: {target_radius_length})")
print(f"- Exact Probability = {prob_radius:.2f} (or {prob_radius * 100:.0f}%)")
```

```python
#Q2
import math

print("=== Q2: Bertrand's Paradox ===\n")

# Given parameters (let's use a unit circle for simplicity)
R = 1


# Method 1: "Random Endpoints"
# Logic: Fix one point. The second point must land on the opposite 1/3 of the circumference.
# Probability = (Target Arc Length) / (Total Circumference)

total_circumference = 2 * math.pi * R
target_arc = total_circumference / 3

prob_method1 = target_arc / total_circumference

print("Method 1: Random Endpoints")
print(f"- Target arc is exactly 1/3 of the circumference.")
print(f"- Exact Probability = 1/3 ≈ {prob_method1:.4f}\n")

# Method 2: "Random Radius"
# Logic: The chord's midpoint must fall on the inner half of the radius [0, R/2].
# Probability = (Target Radius Length) / (Total Radius Length)

total_radius = R
target_radius_segment = R / 2

prob_method2 = target_radius_segment / total_radius

print("Method 2: Random Radius")
print(f"- Midpoint must fall between 0 and R/2 on a radius of length R.")
print(f"- Exact Probability = 1/2 = {prob_method2:.4f}\n")


# Method 3: "Random Midpoint"
# Logic: The chord's midpoint must fall inside a concentric circle of radius R/2.
# Probability = (Area of Inner Circle) / (Area of Total Circle)

total_area = math.pi * (R ** 2)
inner_area = math.pi * ((R / 2) ** 2)

prob_method3 = inner_area / total_area

print("Method 3: Random Midpoint")
print(f"- Midpoint must fall within the inner circle (Area = pi*(R/2)^2).")
print(f"- Total Area = pi*R^2.")
print(f"- Exact Probability = 1/4 = {prob_method3:.4f}")
```

```python
#Q8
# Define the full sample space for a family with two children
# 'B' = Boy, 'G' = Girl
sample_space = [('B', 'B'), ('B', 'G'), ('G', 'B'), ('G', 'G')]

# Event B: "Given that AT LEAST ONE of them is a girl"
# We filter the sample space to keep only the outcomes that have at least one 'G'
condition_space = [family for family in sample_space if 'G' in family]

# Event A: "Both children are girls"
# We look inside our new restricted sample space to find the target event
target_event = [family for family in condition_space if family == ('G', 'G')]

# Calculate the exact conditional probability: P(A|B) = Count(A AND B) / Count(B)
probability = len(target_event) / len(condition_space)

# Print the step-by-step logic and the final result
print("--- Step-by-Step Breakdown ---")
print(f"1. Original Sample Space: {sample_space} (Total: {len(sample_space)})")
print(f"2. Given Condition (At least 1 Girl): {condition_space} (Total: {len(condition_space)})")
print(f"3. Target Event (Both Girls): {target_event} (Total: {len(target_event)})")
print("\n--- Final Answer ---")
print(f"Probability = {len(target_event)} / {len(condition_space)} = {probability:.4f} (or 33.33%)")
```

```python
#Q9

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import combinations

# Defining the core data
omega = ['G1', 'G2', 'D1', 'D2']
P = {'G1': 0.50, 'G2': 0.20, 'D1': 0.10, 'D2': 0.20}

print("=== Q9 Complete Solution ===\n")

# Part 1: Probability Space
print("1. Probability Space (Ω, F, P):")
print(f"   Ω = {omega}")
print(f"   F = P(Ω) (The Power set of Ω, containing all 16 subsets)")
print(f"   P = {P}\n")

# Part 2: Sigma-algebra generated by X
print("2. σ-algebra generated by X, namely σ(X):")
sigma_X = [set(), set(omega), {'G1', 'G2'}, {'D1', 'D2'}]
print(f"   σ(X) = {sigma_X}\n")

# Part 3: Sigma-algebra generated by Y
print("3. σ-algebra generated by Y, namely σ(Y):")
def powerset(iterable):
    s = list(iterable)
    return [set(c) for i in range(len(s) + 1) for c in combinations(s, i)]
sigma_Y = powerset(omega)
print(f"   σ(Y) = Power set containing all {len(sigma_Y)} subsets of Ω.\n")

# Part 4: Show σ(X) ⊆ σ(Y)
print("4. Show σ(X) ⊆ σ(Y) and explain:")
is_subset = all(s in sigma_Y for s in sigma_X)
print(f"   Is σ(X) a subset of σ(Y)? {is_subset}")
print("   Explanation: Y assigns a unique value to every outcome, separating the sample space completely.")
print("   Therefore, its σ-algebra is the entire power set, which inherently contains the coarser partitions of X.\n")

# Part 5: Marginal distribution of X
print("5. Marginal distribution of X:")
P_X_0 = P['G1'] + P['G2']
P_X_1 = P['D1'] + P['D2']
print(f"   P(X=0) = {P_X_0:.2f}")
print(f"   P(X=1) = {P_X_1:.2f}\n")

# Part 6: Marginal distribution of Y
print("6. Marginal distribution of Y:")
print(f"   P(Y=1) = {P['G1']:.2f}, P(Y=2) = {P['G2']:.2f}, P(Y=3) = {P['D1']:.2f}, P(Y=4) = {P['D2']:.2f}\n")

# Part 7: Joint probabilities
print("7. Compute the joint probabilities P(X, Y):")
print(f"   P(X=0, Y=1) = {P['G1']:.2f}")
print(f"   P(X=0, Y=2) = {P['G2']:.2f}")
print(f"   P(X=1, Y=3) = {P['D1']:.2f}")
print(f"   P(X=1, Y=4) = {P['D2']:.2f}\n")

# Part 8: Conditional probabilities
print("8. Compute the conditional probabilities:")
print(f"   P(X=1 | Y=3) = {P['D1'] / P['D1']:.2f}")
print(f"   P(X=1 | Y=4) = {P['D2'] / P['D2']:.2f}")
print(f"   P(Y=3 | X=1) = {P['D1'] / P_X_1:.4f}")
print(f"   P(Y=4 | X=1) = {P['D2'] / P_X_1:.4f}\n")

# Part 9: Interpret the difference between conditioning
print("9. Interpret conditioning on X vs Y:")
print("   - Conditioning on Y: Removes all uncertainty about X. If you know the machine & quality (Y), you know the quality (X).")
print("   - Conditioning on X: Leaves uncertainty about Y. Knowing an item is defective (X=1) doesn't tell you which machine made it.\n")

# Part 10: Information available
print("10. Information available in σ(X) vs σ(Y):")
print("    - σ(X) only contains information about whether the product is Good or Defective.")
print("    - σ(Y) contains information about BOTH the quality and which specific machine produced it.\n")

# Part 12: Shannon Entropy (Printed before the plot so console output is clean)
print("12. Shannon Entropy (Marginal, Joint, and Conditional):")
def entropy(probs):
    p = np.array(probs)
    p = p[p > 0] # Filter out zeros to avoid log(0) error
    return -np.sum(p * np.log2(p))

H_X = entropy([P_X_0, P_X_1])
H_Y = entropy(list(P.values()))
H_XY = H_Y # Since X is a deterministic subset of Y, H(X,Y) = H(Y)
print(f"    Marginal H(X) = {H_X:.4f} bits")
print(f"    Marginal H(Y) = {H_Y:.4f} bits")
print(f"    Joint H(X,Y)  = {H_XY:.4f} bits")
print(f"    Cond H(X|Y)   = {H_XY - H_Y:.4f} bits (Zero uncertainty)")
print(f"    Cond H(Y|X)   = {H_XY - H_X:.4f} bits (Remaining uncertainty)\n")

# Part 11: Python Visualization
print("11. Python visualization generating...")
joint_matrix = np.array([
    [P['G1'], P['G2'], 0, 0],
    [0, 0, P['D1'], P['D2']]
])

plt.figure(figsize=(9, 4))
sns.heatmap(joint_matrix, annot=True, cmap="Blues", cbar=True,
            xticklabels=['Y=1 (G1)', 'Y=2 (G2)', 'Y=3 (D1)', 'Y=4 (D2)'],
            yticklabels=['X=0 (Good)', 'X=1 (Defective)'])
plt.title("Part 11: Joint Probability Distribution P(X, Y)")
plt.xlabel("Random Variable Y")
plt.ylabel("Random Variable X")
plt.show()
```

```python
#Q10
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import math

# Parts 1 to 4: Probability Spaces & Sigma Algebras
print("1. Probability Space (Ω, F, P):")
print("   Ω = R² (the 2D real plane)")
print("   F = B(R²) (the Borel σ-algebra on R²)")
print("   P = The probability measure defined by the joint density f_T,P(t,p).")

print("\n2. σ-algebra generated by X, namely σ(X):")
print("   X only records temperature (T). So σ(X) consists of vertical 'strips' in R².")
print("   σ(X) = { B × R | B ∈ B(R) }, where B is any Borel set on the real line.")

print("\n3. σ-algebra generated by Y, namely σ(Y):")
print("   Y records both (T, P). It separates every point in the plane.")
print("   σ(Y) = B(R²) (the entire Borel σ-algebra on R²).")

print("\n4. Show σ(X) ⊆ σ(Y) and explain:")
print("   Every set in σ(X) is of the form B × R. These are all valid sets in B(R²).")
print("   Thus, σ(X) ⊆ σ(Y). Y contains more information because it tracks the")
print("   pressure deviations, whereas σ(X) is entirely 'blind' to the pressure axis.")

# Parts 5 and 6: Marginal Densities
print("\n--- Marginal Densities ---")
print("5. Marginal density of X:")
print("   Since X(t,p) = t, X is just the random variable T ~ N(0,1).")
print("   f_X(x) = (1 / sqrt(2π)) * exp(-x² / 2)")

print("\n6. Marginal density of Y:")
print("   Since Y(t,p) = (t,p), Y is the joint distribution.")
print("   f_Y(t,p) = f_T,P(t,p) = (1 / 4π) * exp(-(t²/2 + p²/8))")

#  Part 7: Probabilities
print("\n--- Probability Computations ---")
# T ~ N(0,1), P ~ N(0,4) -> standard deviation of P is sqrt(4) = 2
p1 = norm.cdf(0, loc=0, scale=1)
p2 = 1 - norm.cdf(1, loc=0, scale=1)
p3 = p1 # Same as P(X <= 0)
p_T_interval = norm.cdf(1, loc=0, scale=1) - norm.cdf(-1, loc=0, scale=1)
p_P_interval = norm.cdf(2, loc=0, scale=2) - norm.cdf(-2, loc=0, scale=2)
p4 = p_T_interval * p_P_interval

print(f"7. P(X <= 0) = {p1:.4f} (exactly half the normal curve)")
print(f"   P(X > 1)  = {p2:.4f}")
print(f"   P(Y ∈ (-∞,0] × R) = {p3:.4f} (equivalent to P(X <= 0))")
print(f"   P(Y ∈ [-1,1] × [-2,2]) = P(-1<=T<=1) * P(-2<=P<=2)")
print(f"                          = {p_T_interval:.4f} * {p_P_interval:.4f} = {p4:.4f}")

#  Parts 8 to 11: Conditional Distributions & Functions
print("\n--- Conditioning ---")
print("8. Conditional distribution of X given Y=(t,p):")
print("   If you know Y, you know T=t exactly. The distribution of X collapses to a")
print("   single point at t. (Mathematically, it's a Dirac delta function: δ(x-t)).")

print("\n9. Conditional distribution of Y given X=t:")
print("   You know T=t, but P is still completely unknown. Because T and P are independent,")
print("   the conditional distribution is Y = (t, P) where P ~ N(0,4).")

print("\n10. Explain in words:")
print("    - Conditioning on X: We only know the temperature. The pressure is still a random guess.")
print("    - Conditioning on Y: We know both temperature and pressure. Zero uncertainty remains.")

print("\n11. Show X = π1 ◦ Y:")
print("    By definition, Y(t,p) = (t,p). The projection π1 takes (t,p) and returns t.")
print("    Therefore, π1(Y(t,p)) = t = X(t,p). Since X is a measurable function of Y,")
print("    its σ-algebra must be contained within Y's σ-algebra (σ(X) ⊆ σ(Y)).")

#  Parts 13 to 16: Differential Entropy
print("\n--- Differential Entropy ---")
# Formula for diff entropy of Normal dist: h = 0.5 * ln(2 * pi * e * sigma^2)
h_X = 0.5 * np.log(2 * np.pi * np.e * 1)
h_P = 0.5 * np.log(2 * np.pi * np.e * 4)
h_Y = h_X + h_P # Because T and P are independent

print(f"13. Differential entropy of X: h(X) = 0.5 * ln(2πe(1)) = {h_X:.4f} nats")
print(f"14. Differential entropy of Y: h(Y) = h(T) + h(P) = {h_Y:.4f} nats")

print(f"\n15. Conditional differential entropy h(Y|X):")
print(f"    h(Y|X) = h(T,P | T) = h(P) = {h_P:.4f} nats.")
print("    Physical interpretation: If you measure the temperature deviation, the only")
print("    randomness left in the system is the pressure deviation.")

print("\n16. What happens to h(X|Y)?")
print("    In a discrete system, knowing Y means X is known, so H(X|Y) = 0.")
print("    However, continuous differential entropy h(X|Y) approaches -∞ as the")
print("    variance goes to 0 (a Dirac delta function has negative infinite entropy).")
print("    This is why differential entropy is not an absolute measure of information like")
print("    discrete Shannon entropy; it is relative to the coordinate system.")
print("="*60)


# Part 12: Python Visualizations

# Setup grid for plotting
t = np.linspace(-3, 3, 200)
p_grid = np.linspace(-6, 6, 200)
T, P_mesh = np.meshgrid(t, p_grid)

# PDF calculations
pdf_T = norm.pdf(t, 0, 1)
pdf_P = norm.pdf(p_grid, 0, 2)
pdf_Y = norm.pdf(T, 0, 1) * norm.pdf(P_mesh, 0, 2)

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# Plot 1: Joint Density of Y
contour = axes[0].contourf(T, P_mesh, pdf_Y, levels=20, cmap='viridis')
axes[0].set_title('12a. Joint Density of Y=(T,P)')
axes[0].set_xlabel('Temperature Deviation (t)')
axes[0].set_ylabel('Pressure Deviation (p)')
fig.colorbar(contour, ax=axes[0])

# Plot 2: Marginal Density of X
axes[1].plot(t, pdf_T, color='blue', lw=2)
axes[1].fill_between(t, pdf_T, alpha=0.3, color='blue')
axes[1].set_title('12b. Marginal Density of X (Temp only)')
axes[1].set_xlabel('Temperature Deviation (t)')
axes[1].set_ylabel('Density f_X(x)')
axes[1].grid(True, alpha=0.3)

# Plot 3: Sigma Algebra Visualization
# Show vertical strips for sigma(X)
for i in np.linspace(-2.5, 2.5, 6):
    axes[2].axvspan(i, i+0.4, color='blue', alpha=0.2)
# Show fine grid overlay to represent sigma(Y)
for i in np.linspace(-3, 3, 12):
    axes[2].axvline(i, color='gray', linestyle=':', alpha=0.5)
for j in np.linspace(-6, 6, 12):
    axes[2].axhline(j, color='gray', linestyle=':', alpha=0.5)

axes[2].set_title('12c. σ(X) vs σ(Y) Concept')
axes[2].set_xlabel('Temperature (X domain)')
axes[2].set_ylabel('Pressure (Invisible to X)')
axes[2].text(0, 4, 'σ(X): Knows only vertical columns\nσ(Y): Knows exact grid coordinates',
             ha='center', va='center', bbox=dict(facecolor='white', alpha=0.8))
axes[2].set_xlim(-3, 3)
axes[2].set_ylim(-6, 6)

plt.tight_layout()
plt.show()
```
